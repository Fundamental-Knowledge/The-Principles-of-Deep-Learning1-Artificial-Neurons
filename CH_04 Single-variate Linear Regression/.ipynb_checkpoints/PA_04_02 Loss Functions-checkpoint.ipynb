# Chapter.4 Single-variate Linear Regression

## Programming Assignment.4-02 Loss Functions


```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

plt.style.use('seaborn')
np.random.seed(0)
```

```python
x1 = 1
y1 = 3*x1 + 1

th1_range = np.linspace(3 - 3, 3 + 3, 100)
th0_range = np.linspace(1 - 3, 1 + 3, 100)

Th1, Th0 = np.meshgrid(th1_range, th0_range)
loss_funct = np.power(y1 - (Th1*x1 + Th0),2)

fig, ax = plt.subplots(figsize = (8,8))
levels = np.geomspace(np.min(loss_funct)+0.01, np.max(loss_funct), 100)
cmap = cm.get_cmap('Reds_r', lut = len(levels))
ax.contour(Th1, Th0, loss_funct,
          levels = levels, cmap = cmap, alpha = 0.5)
```

```python
def get_loss_funct(x, y, ax):
    th1_range = np.linspace(3 - 3, 3 + 3, 100)
    th0_range = np.linspace(1 - 3, 1 + 3, 100)

    Th1, Th0 = np.meshgrid(th1_range, th0_range)
    loss_funct = np.power(y - (Th1*x + Th0),2)

    levels = np.geomspace(np.min(loss_funct)+0.01, np.max(loss_funct), 100)
    cmap = cm.get_cmap('Reds_r', lut = len(levels))
    ax.contour(Th1, Th0, loss_funct,
              levels = levels, cmap = cmap, alpha = 0.5)
    ax.tick_params(axis = 'both', labelsize = 20)
    ax.set_xlabel(r'$\theta_{1}$', fontsize = 20)
    ax.set_ylabel(r'$\theta_{0}$', fontsize = 20)
```

```python
th1, th0 = 0.1, 0.1
lr = 0.01
iterations = 500

th1_list, th0_list = [], []
loss_list = []

for iteration in range(iterations):
    pred = th1*x1 + th0
    loss = np.power(y1 - pred, 2)

    th1_list.append(th1)
    th0_list.append(th0)
    loss_list.append(loss)
    
    th1 = th1 + 2*x1*lr*(y1 - pred)
    th0 = th0 + 2*lr*(y1 - pred)

fig, ax = plt.subplots(2, 1, figsize = (20,10))
ax[0].plot(th1_list)
ax[0].plot(th0_list)
ax[1].plot(loss_list)
ax[0].set_title(r'$\theta$', fontsize = 30)
ax[1].set_title(r'$\mathcal{L}$', fontsize = 30)
for ax_idx in range(2):
    ax[ax_idx].tick_params(axis = 'both', labelsize = 20)
```

```python
def trainer(iterations, lr, x, y):
    th1, th0 = 0.1, 0.1
    lr = 0.01
    iterations = 500

    th1_list, th0_list = [], []
    loss_list = []

    for iteration in range(iterations):
        pred = th1*x + th0
        loss = np.power(y - pred, 2)

        th1_list.append(th1)
        th0_list.append(th0)
        loss_list.append(loss)

        th1 = th1 + 2*x*lr*(y - pred)
        th0 = th0 + 2*lr*(y - pred)
    return th1_list, th0_list, loss_list
```

```python
x1 = 1
y1 = 3*x1 + 1

th1, th0 = 0.1, 0.1
lr = 0.01
iterations = 500

th1_list, th0_list, loss_list = trainer(iterations, lr, x1, y1)
fig, ax = plt.subplots(2, 1, figsize = (20,10))
ax[0].plot(th1_list)
ax[0].plot(th0_list)
ax[1].plot(loss_list)
ax[0].set_title(r'$\theta$', fontsize = 30)
ax[1].set_title(r'$\mathcal{L}$', fontsize = 30)
for ax_idx in range(2):
    ax[ax_idx].tick_params(axis = 'both', labelsize = 20)
```

```python
    
x1 = 1
y1 = 3*x1 + 1

fig, ax = plt.subplots(figsize = (10,10))

get_loss_funct(x1, y1, ax)
th1_list, th0_list, loss_list = trainer(iterations, lr, x1, y1)

ax.scatter(th1_list, th0_list)
```

```python
x_data = np.array([1, 2, 3])
y_data = 3*x_data + 1

fig, ax = plt.subplots(1, 3, figsize = (30,10))

for data_idx in range(len(x_data)):
    x, y = x_data[data_idx], y_data[data_idx]
    
    get_loss_funct(x, y, ax[data_idx])
    th1_list, th0_list, loss_list = trainer(iterations, lr, x, y)
    ax[data_idx].scatter(th1_list, th0_list)
```
